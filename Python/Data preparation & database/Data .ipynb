## Create a list with all the unique ingredients

```python
# Split the string inside ingredients and add all the ingredients to the list
ing_list =[]
for i in range(len(df_rr['ingredients'])):
    temp = df_rr['ingredients'].loc[i].split(", ")
    
    for j in range(len(temp)):
        ingredient = temp[j]
        ing_list.append(ingredient)
        
```

```python
# Convert list to set and then to list to get all the unique values
unique_ing = list(set(ing_list))
unique_ing = sorted(unique_ing)
str(unique_ing)
        
```

```python
len(unique_ing), len(ing_list)
```

```python
# Count how often each ingredient is used
from collections import Counter
z = ing_list
Counter(z)
```

## Create a list with all the plants that can be grown in the (vertical) garden

```python
plant_list =[]
for i in range(len(df_rr['plants'])):
    temp = df_rr['plants'].loc[i].split(", ")
    
    for j in range(len(temp)):
        plant = temp[j]
        plant_list.append(plant)
unique_plant = list(set(plant_list))
unique_plant = sorted(unique_plant)
len(unique_plant)
```

```python
unique_plant
```

```python
# How often each plant is used as an ingredient in a recipe
y = plant_list
Counter(y)
```

## Create new column with ingredient vectors

```python
df_new = df_rr
df_new['Ingredient vector'] = '0,1'

df_new.head(2)
```

When the ingredient is present a 1 will be added to the string and otherwise a 0

```python
for i in range(len(df_new['ingredients'])):
    vector = [] 
    tempStr = df_new['ingredients'].loc[i].split(", ")
    for j in range(len(unique_ing)):
        if unique_ing[j] in tempStr:  
            vector.append(1)
        else:
            vector.append(0)       
    df_new['Ingredient vector'].loc[i]= vector

df_new.head(2)
```
